## Data loading

Load the "hour.csv" dataset into a pandas DataFrame.


**Reasoning**:
Load the "hour.csv" dataset into a pandas DataFrame and display the first few rows to verify the load.

## Data exploration

### Subtask:
Explore the loaded dataset to understand its structure, identify potential issues, and gain initial insights.


**Reasoning**:
I need to explore the dataset's structure, identify potential issues, and gain initial insights using `.shape`, `.info()`, `.describe()`, `.unique()` or `.value_counts()`, and visualizations.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Get an overview of data types and missing values
print("\nInfo:")
df.info()

# Descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
display(df.describe())

# Explore unique values in categorical columns
categorical_cols = ['season', 'weathersit', 'weekday']
for col in categorical_cols:
    print(f"\nUnique values in '{col}':")
    print(df[col].value_counts())

# Visualize distributions of key numerical features
import matplotlib.pyplot as plt

numerical_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Data cleaning

### Subtask:
Clean the data by handling inconsistencies or potential issues identified during data exploration.


**Reasoning**:
Convert the 'dteday' column to datetime objects, handle potential outliers in numerical features using the IQR method, check for and remove duplicate rows, and ensure categorical columns have the correct data type.



```python
import pandas as pd
import numpy as np

# Convert 'dteday' to datetime
df['dteday'] = pd.to_datetime(df['dteday'])

# Handle outliers in numerical features using IQR
numerical_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Ensure categorical columns have the correct data type
categorical_cols = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']
for col in categorical_cols:
    df[col] = df[col].astype('category')

display(df.head())
display(df.info())
```

## Data preparation

### Subtask:
Prepare the data for the Streamlit dashboard.


**Reasoning**:
Create the `df_dashboard` DataFrame by selecting relevant columns and creating aggregated features. Convert data types to appropriate formats for the dashboard.



```python
# Select relevant columns and create aggregated features
df_dashboard = df[['dteday', 'hr', 'season', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']].copy()

# Aggregate rentals per day
daily_rentals = df_dashboard.groupby('dteday')['cnt'].sum().reset_index()
daily_rentals.rename(columns={'cnt': 'daily_total'}, inplace=True)

# Merge aggregated data with the main DataFrame
df_dashboard = pd.merge(df_dashboard, daily_rentals, on='dteday', how='left')

# Convert 'dteday' to datetime if it isn't already
if not pd.api.types.is_datetime64_any_dtype(df_dashboard['dteday']):
    df_dashboard['dteday'] = pd.to_datetime(df_dashboard['dteday'])

# Ensure data types are suitable
for col in ['season', 'weathersit', 'hr']:
    df_dashboard[col] = df_dashboard[col].astype('category')

display(df_dashboard.head())
display(df_dashboard.info())
```

## Data analysis

### Subtask:
Analyze the data to answer two key business questions.


**Reasoning**:
Define two business questions and analyze the data to answer them, then document the findings.



```python
# Business Question 1: How does the average number of bike rentals vary across different hours of the day?
hourly_rentals = df.groupby('hr')['cnt'].mean().reset_index()

# Business Question 2: Is there a correlation between temperature and the number of bike rentals?
correlation = df['temp'].corr(df['cnt'])

# Documenting the findings
print("Business Question 1: How does the average number of bike rentals vary across different hours of the day?")
display(hourly_rentals)
print("\nThe average number of bike rentals shows a clear pattern across different hours.  Further visualization would help to understand this pattern.")

print("\nBusiness Question 2: Is there a correlation between temperature and the number of bike rentals?")
print(f"The correlation coefficient between temperature and the number of bike rentals is: {correlation}")
print("\nThis indicates a positive correlation between temperature and bike rentals. Higher temperatures tend to be associated with more bike rentals.")
```

**Reasoning**:
Visualize the results of the analysis to better understand the relationship between hour of the day and average bike rentals, and the correlation between temperature and bike rentals.



```python
import matplotlib.pyplot as plt

# Visualization for Business Question 1
plt.figure(figsize=(12, 6))
plt.plot(hourly_rentals['hr'], hourly_rentals['cnt'], marker='o', linestyle='-', color='skyblue')
plt.title('Average Bike Rentals per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Rental Count')
plt.grid(True)
plt.xticks(range(24))
plt.show()

# Visualization for Business Question 2 (Scatter Plot)
plt.figure(figsize=(8, 6))
plt.scatter(df['temp'], df['cnt'], color='salmon', alpha=0.5)  # Use alpha for better visibility with many points
plt.title('Bike Rentals vs. Temperature')
plt.xlabel('Temperature')
plt.ylabel('Rental Count')
plt.grid(True)
plt.show()
```
